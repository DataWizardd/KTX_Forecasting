```python
def eda_basic_info(df, df_name="DataFrame"):
    print(f"\n===== {df_name} =====")
    print("Head (5 rows):")
    display(df.head())
    
    print("\nInfo:")
    print(df.info())
    
    print("\nDescribe (numeric):")
    display(df.describe())
    
    print("\nNull counts:")
    print(df.isnull().sum())
    
    print("="*40)
```

```python
# '운행일자'를 날짜 형식으로 변환
df['운행일자'] = pd.to_datetime(df['운행일자'])

# 역무열차종 드롭
df = df.drop(columns=['역무열차종'])

# 열별 집계 함수 설정
agg_funcs = {
    '공급차량수': 'sum',
    '공급좌석합계수': 'sum',
    '승차수입금액': 'sum',
    '승차인원수': 'sum',
    '승차연인거리': 'sum',
    '좌석거리': 'sum',
    '상행_KTX': 'sum',
    '상행_KTX-산천': 'sum',
    '상행_KTX-호남': 'sum',
    '하행_KTX': 'sum',
    '하행_KTX-산천': 'sum',
    '하행_KTX-호남': 'sum',
}

# 나머지 열을 첫 번째 값으로 유지
for col in df.columns:
    if col not in agg_funcs and col not in ['운행일자', '주운행선']:
        agg_funcs[col] = 'first'

# 그룹화 및 집계
df = df.groupby(['운행일자', '주운행선'], as_index=False).agg(agg_funcs)

# 결과 확인
df
```

```python
# 날짜 기준 정렬
df = df.sort_values(by=['주운행선', '운행일자'])
df
```

```python
# 합계(sum)로 묶을 열 목록
sum_columns = [
    '공급차량수', '공급좌석합계수', '승차수입금액', '승차인원수', 
    '승차연인거리', '좌석거리', '상행_KTX', '상행_KTX-산천', 
    '상행_KTX-호남', '하행_KTX', '하행_KTX-산천', '하행_KTX-호남',
    '공휴일여부', '평일', '주말', '일', '주차'
]

# 그룹화 및 집계 설정
monthly_grouped = df.groupby(['년도', '월', '주운행선'], as_index=False).agg(
    {**{col: 'sum' for col in sum_columns}, **{col: 'first' for col in df.columns if col not in ['년도', '월', '주운행선'] + sum_columns}}
)

# 결과 출력
monthly_grouped
```

```python
# 드랍할 열 목록
drop_columns = [
    '운행요일구분_금', '운행요일구분_목', '운행요일구분_수',
    '운행요일구분_월', '운행요일구분_일', '운행요일구분_토', '운행요일구분_화', '주차'
]

# 열 드랍
monthly_grouped = monthly_grouped.drop(columns=drop_columns)
```

```python
# 드랍할 열 목록
drop_columns = [
    '일','운행일자'
]

# 열 드랍
monthly_grouped = monthly_grouped.drop(columns=drop_columns)
```

```python
# '년도', '월', '주운행선' 기준으로 정렬
monthly_grouped = monthly_grouped.sort_values(['년도', '월', '주운행선'])

# lag 변수 생성 (1, 2, 3개월 전 데이터)
for lag in [1, 2, 3]:
    monthly_grouped[f'승차인원수_lag{lag}'] = monthly_grouped.groupby(['주운행선'])['승차인원수'].shift(lag)
```

```python
# 결측치(NaN)를 0으로 채우기
monthly_grouped = monthly_grouped.fillna(0)
```

```python
df_vix = pd.read_csv(fr"C:\Users\james\J_Data_Lab\Project_KTX\data\KOSPI_Volatility_Index.csv")
df_vix
```

```python
# 날짜 형식 변환
df_vix['날짜'] = pd.to_datetime(df_vix['날짜'])

# 연도-월 단위로 그룹화하여 평균 계산
df_vix['연월'] = df_vix['날짜'].dt.to_period('M')  # 연월 추출
df_vix_monthly = df_vix.groupby('연월')['종가'].mean().reset_index()

# 연월을 문자열로 변환 (YYYY-MM 형식)
df_vix_monthly['연월'] = df_vix_monthly['연월'].astype(str)

# 결과 출력
df_vix_monthly
```

```python
# 연월 변수에서 년도와 월 추출
df_vix_monthly[['년도', '월']] = df_vix_monthly['연월'].str.split('-', expand=True)

# 데이터 타입 변환
df_vix_monthly['년도'] = df_vix_monthly['년도'].astype(int)
df_vix_monthly['월'] = df_vix_monthly['월'].astype(int)

# 연월 변수 삭제
df_vix_monthly.drop('연월', axis=1, inplace=True)

# 결과 확인
df_vix_monthly
```

```python
# 종가 변수 이름 변경
df_vix_monthly.rename(columns={'종가': 'KSVKOSPI'}, inplace=True)

# df와 df_vix_monthly 병합 (년도, 월 기준)
df = pd.merge(df, df_vix_monthly, on=['년도', '월'], how='left')

# 결과 확인
df
```
